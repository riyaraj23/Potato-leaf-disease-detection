```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
from IPython.display import HTML

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import os
import shutil
import numpy as np

source_path = '.' 
# This automatically finds all folders in your current directory 
# and ignores the 'dataset' folder we are creating.
classes = [d for d in os.listdir(source_path) 
           if os.path.isdir(os.path.join(source_path, d)) and d != 'dataset']

print(f"Found these classes: {classes}")

for cls in classes:
    # Create target directories
    for split in ['train', 'val', 'test']:
        os.makedirs(f'dataset/{split}/{cls}', exist_ok=True)
    
    src = os.path.join(source_path, cls)
    all_imgs = [f for f in os.listdir(src) if os.path.isfile(os.path.join(src, f))]
    
    if len(all_imgs) == 0:
        continue
        
    np.random.shuffle(all_imgs)
    
    # Split 70/10/20
    train_idx = int(len(all_imgs) * 0.7)
    val_idx = int(len(all_imgs) * 0.8)
    
    train = all_imgs[:train_idx]
    val = all_imgs[train_idx:val_idx]
    test = all_imgs[val_idx:]
    
    for name, subset in [('train', train), ('val', val), ('test', test)]:
        for img in subset:
            shutil.copy(os.path.join(src, img), f'dataset/{name}/{cls}/{img}')

print("Done! Your data is organized in the 'dataset' folder.")
```

```python
IMAGE_SIZE=256
CHANNELS=3

train_datagen=ImageDataGenerator(
    rescale=1./255,
    horizontal_flip=True,
    rotation_range=10

)
train_generator=train_datagen.flow_from_directory(
    'dataset/Train',
    target_size=(IMAGE_SIZE,IMAGE_SIZE),
    batch_size=32,
    class_mode='sparse',
    
    
)

```
